Ill conditioned and well conditioned 

```python
import numpy as np

# Create a 5x5 Hilbert matrix
n = 5
H = np.array([[1/(i+j+1) for j in range(n)] for i in range(n)])  # Note: i+j+1 for 0-indexing
b = np.ones(n)  # Simple right-hand side

cond_num = np.linalg.cond(H)
print(f"Condition number of 5x5 Hilbert matrix: {cond_num:.2e}")
# Output: Condition number is already ~4.77e+05 for n=5!
# For n=10, it's ~1.60e+13. For n=20, it's ~2.45e+28.
```
