```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define my demand function
def q(x):
    return 20000 - 10000 * x
```

```python
# define my profit function
def profit(x):
    return (x - 0.6 ) * q(x)
```

```python
# feasible region 0.6  to 2
x = np.linspace(0.6,2,400)
```

```python
# Evaluation of the optimal price 
px = profit(x)
```

```python
x_optimal = x[np.argmax(px)]
px_optimal = np.max(px)
```

```python
print("The optimal price is:", x_optimal)
print("The Quantity sold is:", q(x_optimal))
print("The maximum profit is:", px_optimal)
```

```python
plt.plot(x,px)
plt.axvline(x_optimal, linestyle='--')
plt.scatter([x_optimal], [px_optimal], color='blue')
plt.xlabel('Price X (USD)')
plt.ylabel('Monthly Profit P(X) (USD)')
plt.title('Profit Maximazation')
plt.grid()
plt.show()
```
